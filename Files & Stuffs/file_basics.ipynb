File handling refers to the process of performing operations on a file, such as **creating, opening, reading, writing and closing it** through a programming interface.

## **Why Do We Need File Handling?**

1. File Access, File Operations.
2. Data Store, Manage.
3. Recovery, Access, Backup.
4. Programming: Automate.

### **Sample Code(s)**

```python
# import os
f = open('assets/ajhsjahksjh.txt', 'x')

print("Name: ", f.name)
print("Mode: ", f.mode)
print("Is Closed?: ", f.closed)
f.close()
print("Is Closed: ", f.closed)
```

### **Read Stuffs**

```python
f = open('test.txt', 'r')
contents = f.read()
print(contents)
```

### **Reading Between the Lines |||**

#### **readline()**
Reads a single line.

#### **readlines()**
Reads file lines into a list.

```python
f = open('test.txt', 'r')
line = f.readline()
# print(line)
lines = f.readlines()
print(lines)
# print(f.read())
```

### **The Better Way**

```python
for line in open('test.txt', 'r'):
    print(line)
```

##### **Task(s):**

1. Read a file, and count the occurance of a user provided string.
2. Read a file, and count the numbers of characters, alphabets, lowercase letters, uppercase letters, digits, special characters and spaces.
3. Read a file, and count the number of vowels present in each line.

```python
for line in open('test.txt', 'r'):
    # print(len(line)) # +1?
    print(len(line.rstrip("\n")))
```

```python
vowels = 'aeiouAEIOU'
for line in open('test.txt', 'r'):
    for c in line:
        if c in vowels:
            print(c)
        if c.islower():
            print("Lower: "+c)
```

### **with keyword**

```python
text = 'My New Text'
with open('test.txt', 'a') as file:
    file.write(text)
```
