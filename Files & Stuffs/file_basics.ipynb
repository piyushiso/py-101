File handling refers to the process of performing operations on a file, such as **creating, opening, reading, writing and closing it** through a programming interface.

## **Why Do We Need File Handling?**

1. File Access, File Operations.
2. Data Store, Manage.
3. Recovery, Access, Backup.
4. Programming: Automate.

### **Sample Code(s)**

```python
# import os
f = open('assets/ajhsjahksjh.txt', 'x')

print("Name: ", f.name)
print("Mode: ", f.mode)
print("Is Closed?: ", f.closed)
f.close()
print("Is Closed: ", f.closed)
```

### **Read Stuffs**

```python
f = open('test.txt', 'r')
contents = f.read()
print(contents)
```

### **Write Stuffs**

```python
text = 'My New Text'
with open('test.txt', 'a') as file:
    file.write(text)
```
