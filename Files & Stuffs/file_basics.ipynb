##### **Task(s):**

1. Read a file, and count the occurrences of a user provided string.
2. Read a file, and count the numbers of characters, alphabets, lowercase letters, uppercase letters, digits, special characters and spaces.
3. Read a file, and count the number of vowels present in each line.

```python
# Read a file, and count the occurrences of a user provided string.

user_str = input("Enter a string to find: ")
file = open('newFile.txt', 'r')
contents = file.read()
occurrences = contents.count(user_str)
print(f"The string \"{user_str}\" appeared {occurrences} times.")
    
```

```python
for line in open('newFile.txt', 'r'):
    print(len(line))
```

```python
for line in open('newFile.txt', 'r'):
    # print(len(line)) # +1?
    print(len(line.rstrip("\n")))
```

```python
# Read a file, and count the number of characters, alphabets, lowercase, uppercase, digits, special characters and spaces.

chars = 0
alphabets = 0
lowercase = 0
uppercase = 0
digits = 0
special = 0
spaces = 0

for line in open('newFile.txt', 'r'):
    for ch in line:
        if ch != '\n':
            chars += 1
        if ch.isalpha():
            alphabets += 1
            if ch.islower():
                lowercase += 1
            if ch.isupper():
                uppercase += 1
        elif ch.isdigit():
            digits += 1
        elif ch.isspace():
            spaces += 1
        else:
            special += 1
            
print(f"No. of Characters: {chars}")
print(f"Alphabets: {alphabets}")
print(f"Lowercase: {lowercase}")
print(f"Uppercase: {uppercase}")
print(f"Digits: {digits}")
print(f"Special: {special}")
print(f"Spaces: {spaces}")
```

```python
# Read a file, and count the number of vowels present in each line.

vowels = 'aeiouAEIOU'
line_no = 1
for line in open('test.txt', 'r'):
    count = 0
    for ch in line:
        if ch in vowels:
            count += 1
    print(f"Line #{line_no}: {count}")
    line_no += 1
```

### **with keyword**

The **with statement** is used to automatically manage resources like files. It ensures that the file is properly closed after its block of code finishes executing, even if an exception occurs

```python
with open('test.txt', 'r') as file:
    contents = file.read()
    print(contents)
```

Is same as:

```python
file = open('test.txt', 'r')
try:
    contents = file.read()
    print(contents)
# except e:
#     print(e)
finally:
    file.close()
```
