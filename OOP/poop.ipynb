# **Object Oriented Programming (OOP)**

Object Oriented Programming empowers developers to build modular, maintainable and scalable applications. OOP is a way of organizing code that uses objects and classes to represent real-world entities and their behavior. In OOP, object has attributes thing that has specific data and can perform certain actions using methods.

- Organizes code into classes and objects.
- Supports encapsulation to group data and methods together.
- Enables inheritance for reusability and hierarchy.
- Allows polymorphism for flexible method implementation.
- Improves modularity, scalability and maintainability.

## **Class & Objects**
**A Class** is a collection of objects. Classes are blueprints for creating objects. A class defines a set of attributes and methods that the created objects (instances) can have.

**An Object** is an instance of a Class. It represents a specific implementation of the class and holds its own data. It consists: State (Attributes that reflects the properties of an object), Behaviour (Methods that reflects the response of an object to other objects), Identity (Unique name that enables one object to interact with other objects).

````python
class Parent:
    def __init__(self, name):
        self.name = name
    
    def show(self):
        print(self.name)
        
class Child(Parent):
    pass

child = Child("John Cena")
child.show()
````

````python
class GrandParent(Child):
    name = 'Grandparent'
    def __init__(self, year):
        self.year = year

    def show(self):
        print(self.year)

class Parent(GrandParent):
    name = 'Parent'
    def show_parent(self):
        print(super().name)

class Child(Parent):
    name = 'Child'
    def show_parent(self):
        print(super().name)

child = Child(2000)
child.show()
child.show_parent()

parent = Parent(1980)
parent.show()
parent.show_parent()

grandparent = GrandParent(1950)
grandparent.show()
grandparent.show_parent()
````

````python
class Mom:
    def __str__(self):
        return "Mom"
    
class Dad:
    def __str__(self):
        return "Dad"
    
class Child(Mom, Dad):
        
    def __str__(self):
        # return super().__str__()
        return Mom.__str__(self) + ' ' + Dad.__str__(self)
    
child = Child()
print(child)
print(Child.__mro__)    # mro: Method Resolution Order -> States the method's precedence.

````

````python

class Mother:
    def __init__(t, fname, **k):
        super().__init__(**k)
        t.fname = fname
    def show(t):
        print(t.fname)

class Father:
    def __init__(t, lname, **k):
        super().__init__(**k)
        t.lname = lname
    def show(t):
        print(t.lname)

class Child(Mother, Father):
    def __init__(t, fname, lname):
        super().__init__(fname = fname, lname = lname)
    def show(t):
        print(t.fname, t.lname)
        

child = Child(fname = 'John', lname = 'Doe')
child.show()
````

````python
# Method Overloading

class Calc:
    # def add(self, *args):
    #     return sum(args)
    def add(self, a = 0, b = 0, c = 0):
        return a + b + c
    
obj = Calc()
print(obj.add(1, 2))
print(obj.add(1, 2, 3))
````

````python
class Father:
    def name(self):
        print("ABC")
    def who(self):
        print("Father")    
        
class Son(Father):
    def who(self):
        # Overriding who() method of Father class.
        print("Son")
        
        
son = Son()
son.name()  # Calling inherited parent's method.
son.who()   # Calling overridden parent's method.
````

5. **Class Methods & Static Methods**

A class method is a method that receives the class itself as the first argument, conventionally named cls. It can access and modify class-level data and is often used to define factory methods.

A static method is a method that does not receive self or cls automatically. It behaves like a normal function but is placed inside a class because it logically belongs there. It cannot access or modify class or instance data directly.

````python
class Bro:
    # @classmethod
    @staticmethod
    def call():
        print("Calling")
        
Bro.call()
# bro = Bro()
# bro.call()
````

````python
class Nothing:
    @staticmethod
    def callStatic():
        print("Calling Static")
        
    @classmethod
    def callClass(self):
        print("Calling Class")
        
    def callNormal():
        print("Calling Normal")
        
Nothing.callStatic()
Nothing.callClass()
Nothing.callNormal()

# nothing = Nothing()
# nothing.callStatic()
# nothing.callClass()
# nothing.callNormal()
````

6. **Abstract Class & Abstract Methods**

An abstract class is a class that cannot be instantiated on its own and is designed to be a blueprint for other classes.

**An Abstract Base Class (ABC)** defines methods that must be implemented by its subclasses, ensuring that the subclasses follow a consistent structure. ABCs allow you to define common interfaces that various subclasses can implement while enforcing a level of abstraction.

To use abstract methods and class, we need to import and use them as:

```python
from abc import ABC, abstractmethod

class Something(ABC):
    @abstractmethod
    def do_something(self):
        pass

class AnotherThing(Something):
    pass    
```



````python
from abc import ABC, abstractmethod

class Map(ABC):
    @abstractmethod
    def show(self):
        pass

class House(Map):
    def show(self):
        print("SHOWING")

house = House()
house.show()
````

````python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass
    
class Dog(Animal):
    pass

dog = Dog()
dog.sound()
````

````python
class Num:
    def __init__(self, n):
        self.n = n
        
n1 = Num(1)
n2 = Num(2)
print(n1.add(n2))
````

````python
class Num:
    def __init__(self, n):
        self.n = n
        
    def __str__(self):
        return f"{self.n}"
        
    def __add__(self, other):
        return Num(self.n + other.n)
    
    def __sub__(self, other):
        return Num(self.n - other.n)
    
n1 = Num(1)
n2 = Num(2)
n3 = n1 + n2
print(n3)
n3 = n1 - n2
print(n3)
````

#### **Custom Exceptions**

````python
class CustomException(Exception):
    def __init__(self, msg):
        self.msg = msg
        
try:
    text = 'ABC'
    if(text == 'ABC'):
        raise CustomException('Text can not be ABC')
    
except CustomException as ce:
    print(ce)
    
finally:
    print("PROGRAM CONCLUDED")
````

#### **Task(s):**

1. Take **age (int type)** as user input, and raise a **custom Exception (InvalidAgeException)** if the age does not lie between the inclusive range of 18-99.

````python
class InvalidAgeException(Exception):
    def __init__(self, msg):
        self.msg = msg
        
    
try:
    age = int(input("Enter age: "))
    if age < 18 or age > 99:
        raise InvalidAgeException('Age must be between 18 & 99.')
    else:
        print("Valid age.")
except InvalidAgeException as ie:
    print(ie)
    
finally:
    print("Program Concluded.")
````

## **Enumeration (Not enumerate() function)**

The enum module, introduced in Python 3.4, provides a way to create enumerations (enums), which are a set of symbolic names bound to unique, constant values. This improves code readability and maintainability by replacing "magic numbers" with meaningful names.

````python
from enum import Enum, auto

class Colors(Enum):
    RED = 1
    GREEN = 2
    BLUE = auto()
    BLACK = auto()
    
print(Colors.RED)
print(Colors.RED.name)
print(Colors.RED.value)
print(Colors.BLUE.value)
print(Colors.BLACK.value)
print(Colors.HARIYO)
````
