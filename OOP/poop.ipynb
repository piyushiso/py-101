# **Object Oriented Programming (OOP)**

Object Oriented Programming empowers developers to build modular, maintainable and scalable applications. OOP is a way of organizing code that uses objects and classes to represent real-world entities and their behavior. In OOP, object has attributes thing that has specific data and can perform certain actions using methods.

- Organizes code into classes and objects.
- Supports encapsulation to group data and methods together.
- Enables inheritance for reusability and hierarchy.
- Allows polymorphism for flexible method implementation.
- Improves modularity, scalability and maintainability.

## **Class & Objects**
**A Class** is a collection of objects. Classes are blueprints for creating objects. A class defines a set of attributes and methods that the created objects (instances) can have.

**An Object** is an instance of a Class. It represents a specific implementation of the class and holds its own data. It consists: State (Attributes that reflects the properties of an object), Behaviour (Methods that reflects the response of an object to other objects), Identity (Unique name that enables one object to interact with other objects).

## **Four Pillars of OOP**

1. Inheritance
2. Polymorphism
3. Encapsulation
4. Abstraction

### **1. Inheritance**
Allows a class (child class) to acquire properties and methods of a parent class. It supports hierarchical classification and promotes code reuse.

### **2. Polymorphism**
Refers to "Same operation, different behavior". It allows functions or methods with the same name to work differently depending on the type of object they are acting upon.

### **3. Encapsulation**

Bundling of data (attributes) and methods (functions) within a class, restricting access to some components to control interactions. A class is an example of encapsulation as it encapsulates all the data that is member functions, variables, etc.

### **4. Abstraction**

Focuses on "what to do" rather than "how to do it". Abstraction hides the internal implementation details while exposing only the necessary functionality.

## **Doing Stuffs**

````python
class Fan:
    def spin(self):
        print("Spinning")
        
    def stopSpin(self):
        print("Stop Spinning")
        
rahul = Fan()
rahul.spin()
rahul.stopSpin()
````

````python
class Apple:
    def eat(self):
        print("Eating Apple.")
        
    def throw(self):
        print("Throwing Apple.")
        
apple_obj = Apple()
apple_obj.eat()
apple_obj.throw()
````

### **Constructor**

```python
class ClassName:
    def __init__(self):
        pass
```

```python
class ClassName:
    def __init__(this):
        pass
```

````python
class Pet:
    def __init__(this, name):
        this.name = name
        
    def display_name(this):
        print(this.name)
        
pet1 = Pet('Tommy')
pet1.display_name()

pet2 = Pet('Piyush')
pet2.display_name()
````

````python
class City:
    def __init__(this, name=''):
        pass
    
    def set_name(this, name):
        this.name = name
    
        
city = City()
city.set_name("jHJKHakjhsjkahkjsh")
print(city.name)
````

````python
class Student:
    def __init__(this, roll = 0, name = 'N/A'):
        this.roll = roll
        this.name = name

    def details(this):
        print(f"Roll#: {this.roll} | Name: {this.name}")
        
stu_1 = Student(1, "Ram Naam")
stu_1.details()

stu_2 = Student(2)
stu_2.details()

stu_3 = Student(name = 'Hari Bol')
stu_3.details()

stu_0 = Student()
stu_0.details()
````
