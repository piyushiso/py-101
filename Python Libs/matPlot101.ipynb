# **Matplotlib Pyplot**

Matplotlib is an open-source Python library used for creating visualizations. It is widely used in scientific research, data science, and machine learning for transforming numerical data into publication-quality figures. 

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

## **In Action**

````python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
````

## **Let's Start**

### **Plot**

````python
import matplotlib.pyplot as plt

x = [1, 2]
plt.plot(x)
plt.show()
````

````python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [-1, -2, 1]

plt.plot(x, y)
plt.show()
````

#### **Adding Label/Title**

````python
import matplotlib.pyplot as plt

age = [20, 30, 40]
salary = [10000, 20000, 30000]

plt.plot(age, salary)
plt.title("Age & Salary Distribution")
````

````python
plt.plot(age, salary)
plt.title('Age & Salary Distribution')
plt.xlabel('Age')
plt.ylabel('Salary')
````

#### **Adding Marker**

````python
plt.plot(age, salary, marker='.')
````



More @ [W3Schools](https://www.w3schools.com/python/matplotlib_markers.asp)

#### **Adding Linestyle**

````python
plt.plot(age, salary, linestyle='', marker='.')
````

#### **Adding Color**

````python
plt.plot(age, salary, color='red')
plt.show()
plt.plot(age, salary, color='y')
plt.show()
plt.plot(age, salary, color='#EF5500')
plt.show()
````

#### **Combo**

````python
# SHORTHAND NOTATION: marker|line|color
plt.plot(age, salary, 'o:b')
````

### **Also, these...**

````python
age = [20, 21, 23, 22, 21]
people = ['Abe', 'Ram', 'Romeo', 'Hari', 'Om']
plt.plot(age, people)
````

````python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 8]
z = [1, 3, 7]
plt.plot(x, y, z, marker='o')
plt.grid()
plt.show()
````

````python
plt.plot(age, salary)
plt.axis('off')
plt.show()
````

----

## **Grid Lines**

````python
plt.plot(age, salary)
plt.grid()
````

````python
plt.plot(age, salary)
plt.grid(False)
plt.show()
````

````python
plt.plot(age, salary)

plt.grid(axis='x', color='red', linestyle='--', linewidth=1, alpha=0.25)
````

---

## **Subplot**

````python
# subplot(nrows, ncolums, index)
plt.subplot(1, 2, 1)
plt.plot(age, salary)
plt.subplot(2, 2, 2)
plt.plot(age, salary)
plt.show()
````

## **Scatter Plot**

````python
plt.scatter(age, salary)
````

## **Bar Graphs**

````python
plt.bar(age, salary)
````

## **Histogram**


````python
plt.hist(age, age)
plt.show()
````

````python
plt.hist2d(age, age)
plt.show()
````

## **Pie Chart**

````python
plt.pie(age, labels=['Ram', 'Hari', 'Shyam'])
plt.show()
````

## **Box Plot**

````python
plt.boxplot(age)
plt.show()
````
