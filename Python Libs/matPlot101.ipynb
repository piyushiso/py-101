# **Matplotlib Pyplot**

Matplotlib is an open-source Python library used for creating visualizations. It is widely used in scientific research, data science, and machine learning for transforming numerical data into publication-quality figures. 

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

## **In Action**

````python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("assets/demo.jpg")

print("Shape:", img.shape)
print("Min:", img.min(), "Max:", img.max(), "Dtype:", img.dtype)

gray = np.mean(img, axis=2)

if img.max() <= 1.0:
    inverted = 1.0 - img
    bright = np.clip(img + 0.2, 0, 1)
else:
    inverted = 255 - img
    bright = np.clip(img + 50, 0, 255).astype(np.uint8)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(inverted)
plt.title("Inverted")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(bright)
plt.title("Brightened")
plt.axis("off")

plt.show()
````

````python
blue_only = img.copy()
blue_only[:, :, 0] = 0
blue_only[:, :, 1] = 0

plt.imshow(blue_only)
plt.axis("off")
plt.show()
````

````python
red_only = img.copy()
red_only[:, :, 1] = 0
red_only[:, :, 2] = 0

plt.imshow(red_only)
plt.axis("off")
plt.show()
````
