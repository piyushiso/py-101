# **<span style='color:lightblue;'>Pandas</span>**

Pandas is a fundamental, open-source Python library used for **data analysis and manipulation**. It provides powerful and flexible data structures designed to make working with labeled, relational data (like spreadsheets or SQL tables) both easy and intuitive.


## **Two Things**
1. **Series:** 1-Dimensional Data (Single Column)
2. **DataFrame:** 2-Dimensional Data (Table with Rows & Columns)

## **Series**

1. Is Indexed
2. Is Ordered
3. Can store mixed types (but is not prefered)

````python
li = [1, 2, 3]
````

````python
import numpy as np

a = np.array(li)
print(a, type(a), sep='\n')
````

````python
import pandas as pd

a = pd.Series(li)
print(a, type(a), sep='\n')
````

### **Custom Index**

````python
b = pd.Series(li, index=['a', 1, 44.5])
print(b)
````

### **Indexing & Subsetting**

### **Slicing: loc vs iloc**

loc: Label

iloc: Index

````python
import pandas as pd

df = pd.DataFrame({'id': [1, 2, 3], 'names': ['Antu', 'Bandu', 'Chandu']}, index=[-1, 0, 1])

print(df)
print("\n")

print(df.loc[0])
print("\n")

print(df.iloc[0])
````

````python
print(df.loc[0:3])
````

````python
print(df.iloc[0:3])
````

### **Filtering**

### **Merging & Joining Dataframes**

**Merge:** Merges dataframe using a common column.

**Join:** Joins different dataframe independent of the columns.

````python
import pandas as pd

df1 = pd.DataFrame([['A', 20], ['B', 30], ['C', 40]], columns=['name', 'marks'])
df2 = pd.DataFrame({'name': ['A', 'B', 'D'], 'grade': ['A+', 'B+', 'F']})

print(df1, df2, sep='\n')

dfc = df1.merge(df2)
print(dfc)
````

````python
import pandas as pd

df1 = pd.DataFrame({'country': ['Nepal', 'India', 'USA'], 'currency': ['Rupees', 'Rupees', 'Dollar']})
print(df1)

print()

df2 = pd.DataFrame({'country': ['Nepal', 'USA'], 'continent': ['Asia', 'N. America']})
print(df2)

print()

df3 = pd.merge(df1, df2, on='country')
print(df3)
````

````python
import pandas as pd

df1 = pd.DataFrame([[1, 10],[2, 20], [3, 30], [4, 40], [5, 50]], columns=['x', 'y'])
print(df1)

print()

df2 = pd.DataFrame([[11], [22], [33], [44], [55]], columns=['z'])
print(df2)

print()

df3 = df1.join(df2)
print(df3)
````
