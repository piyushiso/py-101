# **Pandas**

Pandas is a fundamental, open-source Python library used for **data analysis and manipulation**. It provides powerful and flexible data structures designed to make working with labeled, relational data (like spreadsheets or SQL tables) both easy and intuitive.


## **Two Things**
1. **Series:** 1-Dimensional Data (Single Column)
2. **DataFrame:** 2-Dimensional Data (Table with Rows & Columns)

**Installation:**

```bash
pip install pandas
```

**Usage:**

```python
import pandas as pd
```

> The pandas library works fabulously with **numpy**.

## **Series**

1. Is Indexed
2. Is Ordered
3. Can store mixed types (but is not prefered)

````python
li = [1, 2, 3]
````

````python
import numpy as np

a = np.array(li)
print(a, type(a), sep='\n')
````

````python
import pandas as pd

a = pd.Series(li)
print(a, type(a), sep='\n')
````

### **Custom Index**

````python
b = pd.Series(li, index=['a', 1, 44])
print(b)
````

### **Dictionary in Series**

````python
country = {'Nepal': 'Bandu', 'India': 'Shyam', 'USA': 'Henry'}
````

````python
import pandas as pd

country_series = pd.Series(country)

print(country_series)
````

````python
import pandas as pd

only_nepal = pd.Series(country, index=['Nepal'])
print(only_nepal)
````

### **Populating Random Data**

````python
import pandas as pd
import numpy as np

series = pd.Series(np.random.randint(low=1, high=6, size=5))
print(series)
````

## **Dataframe**

2D Representation

````python
data = {
    "name": ["Ram", "Sita", "Hari"],
    "marks": [80, 90, 70]
}

df = pd.DataFrame(data)
print(df, type(df), sep='\n')
````

````python
import pandas as pd
import numpy as np

data = [['Bandu', 'USA'], ['Baklol', 'India'], ['Baka', 'Japan']]
df = pd.DataFrame(data, columns=['name', 'country'])
print(df)
````

````python
print(df[0])
print(df[0][0])
````

````python
print(df.head(1))
````

````python
print(df.tail(1))
````

### **More Attributes**
