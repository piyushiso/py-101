## **Array Attributes**

|Attributes|	Description|
|---|---|
|ndim	|returns number of dimension of the array|
|size	|returns number of elements in the array|
|dtype	|returns data type of elements in the array|
|shape	|returns the size of the array in each dimension.|
|itemsize	|returns the size (in bytes) of each elements in the array|
|data	|returns the buffer containing actual elements of the array in memory|


````python

import numpy as np

arr = np.array([1, 2, 3])
print(arr.shape)

arr = np.array([[1], [2]])
print(arr.shape)
````

````python
import numpy as np

def display(arr: np.array):
    print(arr, f"N Dimension: {arr.ndim}", f"Size: {arr.size}", f"dType: {arr.dtype}", f"Item Size: {arr.itemsize}", f"Shape: {arr.shape}", f"Data: {arr.data}", '\n', sep='\n')
    
arr = np.array([1, 2, 3])   # One Dimensional
display(arr)

arr = np.array([[1, 2, 3]]) # Two Dimensional
display(arr)

arr = np.array([[1, 2, 3], [4, 5, 6]])   # Two Dimensional
display(arr)

arr = np.array([[[1], [2]]])   # Three Dimensional
display(arr)
````

## **Copying**

````python
import numpy as np

a = np.array([1, 2, 3])
b = a.view()
print(a, b)
b[0] = 200
print(a, b)
````

````python
import numpy as np

a = np.array([1, 2, 3])
b = a.copy()
print(a, b)
b[0] = 200
print(a, b)
````

````python
import numpy as np

a = np.array([[1], [2]])
b = a.copy()

print(a, b, sep='\n')
print('\n')

b[0][0] = 200

print(a, b, sep='\n')
````

### **searchsorted()**

If you have a sorted array and want to find the index where a specific value would be inserted to maintain the sort order (binary search), use numpy.searchsorted(). This is much faster than a linear search for large arrays.

````python
import numpy as np

arr = np.array([0, 1, 2, 3, 4, 5])
print(arr)
print(arr.searchsorted(2))

# print(arr.searchsorted(-2))


# arr = np.insert(arr, 2, 7)
# print(arr)
````

#### **Task**

Write a code to display mean, variance, and standard deviation of the values of a .csv file.

````python
import numpy as np

data = []

with open("assets/data.csv", "r") as f:
    for line in f:
        values = line.strip().split(",")
        for v in values:
            data.append(float(v))

arr = np.array(data)

print("Mean:", np.mean(arr))
print("Variance:", np.var(arr))
print("Standard Deviation:", np.std(arr))
````

````python
import csv
import numpy as np

data = []

with open("assets/data.csv", "r") as f:
    reader = csv.reader(f)

    for row in reader:
        for value in row:
            data.append(float(value))

arr = np.array(data)

print("Mean:", np.mean(arr))
print("Variance:", np.var(arr))
print("Standard Deviation:", np.std(arr))
````

## **Array Broadcasting**

Broadcasting in NumPy allows us to perform arithmetic operations on arrays of different shapes without reshaping them. It automatically adjusts the smaller array to match the larger array's shape by replicating its values along the necessary dimensions. This makes element-wise operations more efficient by reducing memory usage and eliminating the need for loops.

When comparing shapes:

> Start from the rightmost dimension and move left.

Two dimensions are compatible if:

1. They are equal, OR
2. One of them is 1

````python
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[1, 2]])

print(a, b, sep='\n')

print(a+b)
````
