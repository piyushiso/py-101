### **<span style='font-size: 26px'> Packing & Unpacking Arguments</span>**

#### **\*args: Uses Tuples**

##### **Packing Arguments**

````python
def add(*nums):
    return sum(nums)

print(add(1, 2))
print(add(1, 2, 2))
print(add(1, 2, 3))
print(add(1))
````

##### **Unpacking Arguments**

````python
def add(a, b, c):
    return a + b + c

num = (1, 2, 3)
print(add(*num))
````

#### **\*\*kwargs: Uses Dictionary**

##### **Packing Arguments**

````python
def display_details(**details):
    print("Details: ", details)

display_details(name='Piyush', age=25, city='Hetauda')
````

##### **Unpacking Arguments**

````python
def display_details(name, age, city):
    print("Details: ", name, age, city)
    
details = {
    'name': 'Piyush',
    'age': 25,
    'city': 'Hetauda'
}

display_details(**details)
````

<span style='font-size: 20px;'>**LeetCode #762:** Prime Number of Set Bits in Binary Representation. <a href='https://leetcode.com/problems/prime-number-of-set-bits-in-binary-representation/description/'>Link</a></span>

<br/>

Given two integers left and right, return the count of numbers in the inclusive range [left, right] having a prime number of set bits in their binary representation.

Recall that the number of set bits an integer has is the number of 1's present when written in binary.

For example, 21 written in binary is 10101, which has 3 set bits.

<pre>

<b>Example 1:</b>

Input: left = 6, right = 10

Output: 4

Explanation:

6  -> 110 (2 set bits, 2 is prime)
7  -> 111 (3 set bits, 3 is prime)
8  -> 1000 (1 set bit, 1 is not prime)
9  -> 1001 (2 set bits, 2 is prime)
10 -> 1010 (2 set bits, 2 is prime)
4 numbers have a prime number of set bits.

<b>Example 2:</b>

Input: left = 10, right = 15

Output: 5

Explanation:

10 -> 1010 (2 set bits, 2 is prime)
11 -> 1011 (3 set bits, 3 is prime)
12 -> 1100 (2 set bits, 2 is prime)
13 -> 1101 (3 set bits, 3 is prime)
14 -> 1110 (3 set bits, 3 is prime)
15 -> 1111 (4 set bits, 4 is not prime)
5 numbers have a prime number of set bits.

Constraints:

1 <= left <= right <= 10^6
0 <= right - left <= 10^4

</pre>

````python
def checkPrime(n):
    if(n <= 1):
        return False
    for i in range(2, n//2):
        if(n % i == 0):
            return False
    return True

def countPrimeSetBits(left, right):
    count = 0
    for _ in range(left, right + 1):
        in_bin = format(_, 'b')
        set_bits = in_bin.count('1')
        # print(in_bin, set_bits, sep=': ')
        if(checkPrime(set_bits)):
            count += 1
    return count

print(countPrimeSetBits(1, 10))
print(countPrimeSetBits(4, 18))
````

It is guaranteed that you can always reach one for all test cases.

Step 3) 7 is odd, add 1 and obtain 8.

Step 4) 8 is even, divide by 2 and obtain 4.  

Step 5) 4 is even, divide by 2 and obtain 2. 

Step 6) 2 is even, divide by 2 and obtain 1.  

<br/>

**Example 2:**

**Input:** s = "10"

**Output:** 1

**Explanation:** "10" corresponds to number 2 in their decimal representation.

Step 1) 2 is even, divide by 2 and obtain 1.  

<br/>

**Example 3:**

Input: s = "1"
Output: 0
 
<br/>

**Constraints:**

1 <= s.length <= 500
