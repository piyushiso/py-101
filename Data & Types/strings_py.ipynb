# **Strings in Python**

## What is a string?

A **string** is a data type used to represent text, it is basically a **sequence/collection of characters**.

A string is an **ordered sequence of characters** enclosed in quotes.

## Strings in Python

### Datatype

```python

str()
```

### Example in Python

```python
s = "Hello World!"
```

````python
# A simple example to create and display a string.

s = 'Hello World!'
print(s)
````

### Key Properties

- **Ordered**: A string keeps characters in a fixed order. 
    
    *Example*: 
    
    "Python" is always "Python" and does not become "hnoPty" until changed.
    ```python
    s = 'Hello World!'
    print(s)
    ```


- **Indexed**: Every character in a string has a position number starting from 0. 
    
    *Example*: 
    
    "Python", [0] => 'P', [1] => 'y', [2] => 't', ...
    ```python
    s = 'Hello World!'
    print(s[0], s[1], s[2])
    ```

````python
s = 'Hello World!'
print(s[0], s[1], s[2])
````

- **Iterable**: It can be operated using a loop, hence iterable. 
    
    *Example*: 
    ```python
    for ch in "Python":
        print(ch)
    ```

````python
for ch in "Python":
    print(ch)
````

- **Immutable**: You cannot change the object after it is created.
    
    *Example*:
    ```python
    # Incorrect Code
    s = "Python"
    s[0] = "J"

    # Correct Approach
    s = "J" + s[1:]
    print(s)
    ```

````python
# Incorrect Code
s = "Python"
s[0] = "J"
````

````python
# Correct Approach
s = "J" + s[1:]
print(s)
````

## Indexing & Slicing

### Indexing

```python
s = "Python"
s[0]    # 'P'
s[3]    # 'h'
s[-1]   # 'n'
```

> -1 means last character

````python
s = "Python"
print(s[0])    # 'P'
print(s[3])    # 'h' 
print(s[-1])   # 'n'
print(s[50])   # Error
````

### Indexing with Loops

#### Using len() function.

```python
s = "Hello!"

for i in range(len(s)):
    print(s[i], end="")
```

````python
s = "Hello!"
print(f"Length of \"{s}\": {len(s)}")

for i in range(len(s)):
    print(s[i], end="")
````

### Slicing

An operation that allows you to extract portions of sequences in a string.

```python
s[start : stop : step]
```

````python
s = "Python"

print(s[0:4])   # 'Pyth'
print(s[:3])    # 'Pyt'
print(s[2:])    # 'thon'
print(s[0:6:2])
````

1. Uppercae
